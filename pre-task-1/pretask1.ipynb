Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Exploring 1-D Gaussian Distributions and visualizing impact on distribution with respect to changes in mean and standard deviation


```python
def plot_1d_gaussian(mean, std_dev, title_suffix):
    x = np.random.normal(mean, std_dev, 1000)
    plt.hist(x, bins=30, density=True, alpha=0.6, color='b')
    pdf = stats.norm.pdf(x, mean, std_dev)

    plt.figure(figsize=(10, 6))
    plt.hist(x, bins=30, density=True, alpha=0.6, color='b')
    plt.plot(x, pdf, label='PDF', color='red')
    plt.title(f'1D Gaussian Distribution {title_suffix}')
    plt.xlabel('X-axis')
    plt.ylabel('Probability Density')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
mean1d1 = 0
std_dev1d1 = 1
plot_1d_gaussian(mean1d1, std_dev1d1, '(Mean 0, Std Dev 1)')
```

```python
mean1d2 = 2
std_dev1d2 = 1
plot_1d_gaussian(mean1d2, std_dev1d2, '(Mean 2, Std Dev 1)')
```

```python
mean1d3 = 0
std_dev1d3 = 2
plot_1d_gaussian(mean1d3, std_dev1d3, '(Mean 0, Std Dev 2)')
```

Exploring 2-D Gaussian Distributions and visualizing impact on distribution with respect to changes in mean and covariance matrix

```python
def plot_2d_gaussian(mean, cov, title_suffix):
    data = np.random.multivariate_normal(mean, cov, 1000)

    plt.figure(figsize=(10, 6))
    plt.scatter(data[:, 0], data[:, 1], alpha=0.5)
    plt.title(f'Scatter Plot of 2D Gaussian Data {title_suffix}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axis('equal')
    plt.grid()
    plt.show()

    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    pos = np.dstack((X, Y))
    rv = stats.multivariate_normal(mean, cov)
    Z = rv.pdf(pos)

    plt.figure(figsize=(10, 6))
    plt.contourf(X, Y, Z, levels=20, cmap='viridis')
    plt.colorbar(label='Probability Density')
    plt.title(f'Contour Plot of 2D Gaussian Distribution {title_suffix}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.axis('equal')
    plt.show()

    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    ax.set_title(f'3D Surface Plot of 2D Gaussian Distribution {title_suffix}')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Probability Density')
    plt.show()
```

```python
mean1 = [0, 0]
cov1 = [[1, 0], [0, 1]]
plot_2d_gaussian(mean1, cov1, '(Mean [0, 0], Covariance [[1, 0], [0, 1]])')
```

```python
mean2 = [0, 0]
cov2 = [[1, 0.8], [0.8, 1]]
plot_2d_gaussian(mean2, cov2, '(Mean [0, 0], Covariance [[1, 0.8], [0.8, 1]])')
```

```python
mean3 = [2, 2]
cov3 = [[1, 0], [0, 1]]
plot_2d_gaussian(mean3, cov3, '(Mean [2, 2], Covariance [[1, 0], [0, 1]])')
```

```python
mean4 = [0, 0]
cov4 = [[2, 0], [0, 1]]
plot_2d_gaussian(mean4, cov4, '(Mean [0, 0], Covariance [[2, 0], [0, 1]])')
```

```python
mean5 = [0, 0]
cov5 = [[1, 0.5], [0.5, 1]]
plot_2d_gaussian(mean5, cov5, '(Mean [0, 0], Covariance [[1, 0.5], [0.5, 1]])')
```
