```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Part 1: Generate 1D Samples from Different Distributions
num_samples = 1000
uniform_samples = np.random.uniform(-1, 1, num_samples)
normal_samples = np.random.normal(0, 1, num_samples)
exponential_samples = np.random.exponential(1, num_samples)
laplace_samples = np.random.laplace(0, 1, num_samples)
```

```python
# Plot histograms
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(uniform_samples, bins=30, density=True, alpha=0.6, color='b')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(2, 2, 2)
plt.hist(normal_samples, bins=30, density=True, alpha=0.6, color='r')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(2, 2, 3)
plt.hist(exponential_samples, bins=30, density=True, alpha=0.6, color='g')
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(2, 2, 4)
plt.hist(laplace_samples, bins=30, density=True, alpha=0.6, color='y')
plt.title('Laplace Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# Part 2: Compare Distributions
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.uniform.pdf(x), label='Uniform', color='b')
plt.plot(x, stats.norm.pdf(x), label='Normal', color='r')
plt.plot(x, stats.expon.pdf(x), label='Exponential', color='g')
plt.plot(x, stats.laplace.pdf(x), label='Laplace', color='y')
plt.title('Comparison of Different Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

```python
# Part 3: Multidimensional Sampling
num_samples = 1000
gaussian_samples = np.random.normal(0, 1, (num_samples, 2))
uniform_samples = np.random.uniform(-1, 1, (num_samples, 2))

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(gaussian_samples[:, 0], gaussian_samples[:, 1], alpha=0.5)
plt.title('2D Gaussian Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(1, 2, 2)
plt.scatter(uniform_samples[:, 0], uniform_samples[:, 1], alpha=0.5)
plt.title('2D Uniform Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
